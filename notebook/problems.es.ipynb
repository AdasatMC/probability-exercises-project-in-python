## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
# TODO
import numpy as np

# número de simulaciónes

n_simul = 1000

# simulación de lanzamientos

d1 = np.random.randint(1,7, n_simul)
d2 = np.random.randint(1,7, n_simul)

# suma de los dados

sum = d1 + d2

# condición

condicion = (sum > 7) | (sum %2 == 0)

# probabilidad

probabilidad = np.sum(condicion) / n_simul



print(f'La probabilidad estimada es: {probabilidad}')



```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
import numpy as np

# Crear la caja de bolas como un diccionario
ball_box = {}
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

# Convertir el diccionario en una lista de colores
colores = list(ball_box.values())

# Número de simulaciones
num_simulaciones = 1000

# Inicializar contadores para eventos exitosos
contador_3_blancas_2_rojas = 0
contador_mismo_color = 0

# Simulación
for _ in range(num_simulaciones):
    # Tomar 5 bolas al azar con reemplazo
    bolas_tomadas = np.random.choice(colores, 5)
    
    # Contar colores
    num_blancas = np.sum(bolas_tomadas == "White")
    num_rojas = np.sum(bolas_tomadas == "Red")
    num_verdes = np.sum(bolas_tomadas == "Green")

    # Verificar condiciones
    if num_blancas == 3 & num_rojas == 2:
        contador_3_blancas_2_rojas += 1
    if num_blancas == 5 | num_rojas == 5 | num_verdes == 5:
        contador_mismo_color += 1

# Calcular probabilidades
probabilidad_3_blancas_2_rojas = contador_3_blancas_2_rojas / num_simulaciones
probabilidad_mismo_color = contador_mismo_color / num_simulaciones

# Mostrar resultados
print(f'Probabilidad de 3 bolas blancas y 2 rojas: {probabilidad_3_blancas_2_rojas}')
print(f'Probabilidad de todas las bolas del mismo color: {probabilidad_mismo_color}')

```
